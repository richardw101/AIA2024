## 1.1) Brief Introduction to tf-idf (Follow)
In information retreival [tf-idf](https://en.wikipedia.org/wiki/Tf%E2%80%93idf) (*term frequency â€“ inverse document frequency*) is a metric that represents how "important" a word is to a corpus of text. While we won't go into detail about how it works, essentially all you need to know is that it balances two metrics.

**Term Frequency**: Is exactly what one would expect, it is the the frequency at which a word is present in a corpus of text. A word with a higher term-frequency score appears much more in the corpus compared to one with a low term frequency.

**Inverse Document Frequency**: If we only used term frequency, common words like "the" or "and" would have a high score, even though they don't give us that much information since they are present in every document. *Inverse Document Frequency* is a metric of how much "information" a word provides, and if a word is common or rare across all documents.

Now that you're aware of what tf-idf scores are, **discuss with your group on how you could use tf-idf to build a KNN classifier**:
1. If the data objects (rows) each represent a tweet, what would the features (columns) of the dataset be?
In addition to the features needed to identify a unique tweet, the remaining feature (columns) would be the tf_idf score for all unique words across all tweets.
2. How would you represent a single tweet as a vector of numbers (values for each of the features)? (**see below**)

   What would it mean to have a value of 0 for a given feature? A zero would mean that that word is not present in the tweet that we are looking at in the particular row. 
   
   What about 0.4? A value of 0.4 for a given feature in the context of TF-IDF scores means that the corresponding word has a moderate relevance in the tweet.

TF-IDF scores range from 0 to 1. A score of 0 means that the word does not appear in the tweet at all, while a score of 1 would mean that the word is highly relevant and unique to that tweet within the given corpus of tweets.

So, a score of 0.4 indicates that the word appears in the tweet with a moderate frequency and is somewhat unique across all tweets. It's not the most defining word in the tweet, but it still carries some significance.

3. What would it mean for two sentences to be "nearest neighbors"? In the context of a K-Nearest Neighbors (KNN) classifier and TF-IDF vectorization, two sentences being "nearest neighbors" means that they are very similar in terms of their content.

When we represent sentences (or tweets in this case) as vectors using TF-IDF scores, the geometric distance between these vectors in the multi-dimensional space corresponds to the similarity of the sentences. The smaller the distance, the more similar the sentences are.

So, if two sentences are "nearest neighbors", it means that the distance between their vector representations is among the smallest compared to distances with other sentences. This implies that these two sentences share a lot of common words with similar TF-IDF scores, i.e., they have similar content.


4. Would you expect two near neighbors to have similar sentiment? Why or why not?
Yes, in general, you would expect two near neighbors to have similar sentiment. This is because the K-Nearest Neighbors (KNN) algorithm works on the principle of similarity, meaning that more similar instances (in this case, tweets) are considered "near" to each other.

When we represent tweets as vectors using TF-IDF scores, similar tweets (those with similar words and thus similar TF-IDF scores) will be closer in the vector space. If the words contributing to this similarity are words that carry sentiment (like 'good', 'bad', 'happy', 'sad', etc.), it's likely that these tweets will also have similar sentiment.

However, it's important to note that this isn't always the case. Sentiment analysis is a complex task and depends not just on individual words, but also on their context. Two tweets could use similar words but in different contexts, leading to different sentiments. Therefore, while generally true, it's not a guaranteed outcome.

For example, here are two sentences that use similar words but in different contexts, leading to different sentiments:

"I am happy that I passed the exam."
"I am not happy that I passed the exam without studying."
Both sentences use the words 'happy', 'I', 'passed', 'the', 'exam'. However, the first sentence has a positive sentiment because the speaker is happy about passing the exam. The second sentence, despite using the word 'happy', has a negative sentiment because the speaker is unhappy that they passed without studying, implying they wanted a challenge or they feel undeserving.


A single tweet can be represented as a vector of numbers using the TF-IDF (Term Frequency-Inverse Document Frequency) method. Here's how it works:

1. **Preprocessing**: The tweet text is preprocessed to remove any noise. This includes removing special characters, converting all text to lower case, removing stop words (common words like 'is', 'the', 'and', etc.), and stemming/lemmatization (reducing words to their root form).

2. **Term Frequency (TF)**: For each word in the preprocessed tweet, calculate the Term Frequency. This is the number of times a word appears in the tweet divided by the total number of words in the tweet.

3. **Inverse Document Frequency (IDF)**: This is the log of the number of tweets divided by the number of tweets that contain the word. Inverse Document Frequency diminishes the weight of words that occur very frequently and increases the weight of words that occur rarely.

4. **TF-IDF**: The TF-IDF score for a word is the product of its TF and IDF scores. 

5. **Vector Representation**: Create a vector where each dimension corresponds to a unique word in the corpus of tweets. The value in each dimension is the TF-IDF score of the word for the given tweet. If a word does not appear in the tweet, its TF-IDF score is 0.

This vector of TF-IDF scores represents the tweet in the feature space. Other features like Tweet ID, User ID, Timestamp, etc., can also be included in the vector as additional dimensions.

You're given some code to calculate Adaboost by hand below, but for it to work, you need to complete the following helper functions.

In `calculate_alpha`, you should calculate alpha for the round, based on the classifier's predictions and the weights of that round. The basic steps are:
1. Calculate the **weighted** error $\epsilon$ for the round by comparing those predictions to y. Remember, the error is the sum of the weights of misclassified instances.
2. Calculate alpha using the following formula:

$$\alpha = 0.5*ln(\frac{1 - \epsilon}{\epsilon})$$

```python
def calculate_alpha(predictions, weights, y):
    """
    Inputs:
        predictions: the predictions (0 or 1) of the classifier in this round.
        weights: the weights of each instance at the end of last round
        y: the class labels (0 or 1) for the instances
    Output: the alpha value for this round
    """
    # BEGIN SOLUTION
    # epsilon is the sum of the weights of the misclassified instances - see above
    epsilon = np.sum(weights[y != predictions])
    # recall that we imported the math library at the beginning of the notebook - hint hint
    alpha = 0.5 * math.log((1 - epsilon) / epsilon)
    return alpha
```

```python
# To test calculate_alpha, we need to make a decision tree and some starting weights
dt = DecisionTreeClassifier(criterion = "gini", max_depth=1, random_state=123).fit(X, y)
predictions = dt.predict(X)
# define the starting weights as 1/n where n is the number of instances
weights = np.repeat(1 / len(X), len(X))

# Alpha should be ~0.84
alpha = calculate_alpha(predictions, weights, y)
print("Alpha:\n", alpha)
np.testing.assert_almost_equal(alpha, 0.8416209435087314)
#alpha
```

Why are we implementing it by hand? This way we can keep track of which samples were chosen each round to visualize them.

When you run the following code, you can see the alpha values for each round, changing as the weights, and therefore the samples, change.

```python
# Let's do 20 rounds of adaboost
rounds = 20
# N is the length of training data
N = len(X)
# Each boostrapped sample will b size N/5
sample_size = N // 5

# Each round, keep track of:
# Which instances (indices) were samples
sample_indices = []
# The classifier built on that sample
dtrees = []
# The alpha value for each that round
alphas = []

# Start our weights off each as 1/N
weights = np.repeat(1 / N, N)

for i in range(rounds):
    indices = np.random.choice(range(N), sample_size, replace=True, p=weights)
    sample_X = X[indices,:]
    sample_y = y[indices]
    # Train a simple decision tree on the sample
    dt = DecisionTreeClassifier(criterion = "gini", max_depth=1).fit(sample_X, sample_y)
    predictions = dt.predict(X)
    
    # Calculate alpha
    alpha = calculate_alpha(predictions, weights, y)
    print(f'Round {i} Alpha: {alpha}')
    
    # Update weights
    weights = update_weights(predictions, weights, y, alpha)
    weights /= sum(weights)
    
    sample_indices.append(indices)
    alphas.append(alpha)
    dtrees.append(dt)
```

```python
# This function makes predictions for a given list of X features by taking the weighted sum of
# its constituent classifiers' predictions.
def boosting_predict(X_test):
    return np.mean([(dtrees[i].predict(X_test) - 0.5) * alphas[i] for i in range(len(dtrees))], axis=0) + 0.5
```

The following code will plot the predictions from individual rounds of adaboost. Note:
* The background color indicates the prediction.
* The faded dots were *not* sampled in this round (e.g. because of their low weights).

```python
# Here is a provided method that will plot the decision boundries of each stump

def plot_predictions(pred_function, sample_indices, subplot=plt, continuous=False):
    plot_colors = "br"
    plot_step = 0.02
    class_names = "AB"

    #

    # Plot the decision boundaries
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    Z = pred_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    if not continuous:
        Z = np.round(Z)
        cs = subplot.contourf(xx, yy, Z, cmap=plt.cm.Paired)
    else:
        cs = subplot.contourf(xx, yy, Z)

    # Plot the training points
    for i, n, c in zip(range(2), class_names, plot_colors):
        idx = np.where(y == i)
        in_sample_idx = np.intersect1d(idx, sample_indices)
        subplot.scatter(X[in_sample_idx, 0], X[in_sample_idx, 1],
                    c=c, cmap=plt.cm.Paired,
                    s=20, edgecolor='k',
                    label="Class %s (Sampled)" % n)
        out_sample_idx = np.setdiff1d(idx, in_sample_idx)
        subplot.scatter(X[out_sample_idx, 0], X[out_sample_idx, 1],
                    c=c, cmap=plt.cm.Paired,
                    s=20, edgecolor='k', alpha=0.1,
                    label="Class %s (Not Sampled)" % n)
```

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np

class MyAdaBoost:
    def __init__(self, n_clf=100):
        self.n_clf = n_clf

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize weights to 1/N
        w = np.full(n_samples, (1 / n_samples))

        self.clfs = []
        self.clf_weights = []
        self.errors = []
        self.alphas = []

        for _ in range(self.n_clf):
            clf = DecisionTreeClassifier(max_depth=1, max_leaf_nodes=2)
            clf.fit(X, y, sample_weight=w)

            y_pred = clf.predict(X)

            # Calculate error
            error = w[(y_pred != y)].sum() / w.sum()
            self.errors.append(error)

            # Calculate alpha
            alpha = 0.5 * np.log((1 - error) / error)
            self.alphas.append(alpha)

            # Update weights
            w *= np.exp(-alpha * y * y_pred)
            w /= np.sum(w)

            self.clfs.append(clf)
            self.clf_weights.append(w.copy())

    def predict(self, X):
        clf_preds = [alpha * clf.predict(X) for alpha, clf in zip(self.alphas, self.clfs)]
        y_pred = np.sum(clf_preds, axis=0)
        return np.sign(y_pred)
```
